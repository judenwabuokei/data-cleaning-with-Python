### Importing the messy dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv('messy_data.csv',sep=';', encoding='latin-1')

# to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df.head()
```

### DROP COLUMN 1 AND 9

```python
df.drop(columns=['Unnamed: 8','IMBD title ID'], inplace=True) 
df
```

### CHANGE THE COLUMNS TITLE TO SMALL CASE AND RENAME SOME COLUMNS

```python
df.columns = df.columns.str.lower().str.replace('original titlê','title').str.replace('genrë¨','genre').str.strip().str.replace(' ','_')
df
```

### CLEANING release_year COLUMN

```python
df.release_year = pd.to_datetime(df.release_year, errors='coerce')
```

```python
df.info()
```

### CLEANING genre COLUMN

```python
df.genre = df.genre.str.split(',', expand = True)[0]
df.genre = df.genre.fillna('No Genre')
```

```python
df
```

### CLEANING duration COLUMN 

```python
df.duration.unique()
```

```python
df.duration = df.duration.str.replace('178c','178').str.replace('Nan','0').str.replace('Not Applicable','0').str.replace('Inf','0')
df.duration = df.duration.str.replace('-','0').str.replace(' ','0')
df.duration = df.duration.fillna('0')
df.duration = df.duration.astype(int)
```

```python
df.info()
```

### CLEANING country COLUMN

```python
df.country.unique()
```

```python
df.country = df.country.str.replace('USA', 'US').str.replace('US.', 'US').str.replace('US', 'USA')
df.country = df.country.str.replace('New Zesland', 'New Zealand').str.replace('New Zeland', 'New Zealand')
df.country = df.country.str.replace('Italy1', 'Italy')
df.country = df.country.fillna('No name')
```

```python
df
```

### CLEANING content_rating COLUMN

```python
df.content_rating.unique()
```

```python
df.content_rating = df.content_rating.str.replace('Unrated','Not Rated')
df.content_rating = df.content_rating.fillna('Not Rated')
```

### CLEANING income COLUMN

```python
df.income.unique()
```

```python
df.income = df.income.str.replace('4o8,035,783', '408,035,783').str.replace(',','').str.lstrip('$ ')
df.income = df.income.fillna('0')
df.income = df.income.astype(np.int64)
```

```python
df.info()
```

### CLEANING votes COLUMN

```python
df.votes.unique()
```

```python
df.votes = df.votes.str.strip().str.replace('.','').fillna('0').astype(np.int64)
```

```python
df.info()
```

### CLEANING score COLUMN

```python
df.score.value_counts()
```

```python
df.score = df.score.replace({'8,6':'8.6','8,7e-0':'8.7','8.7.':'8.7','++8.7':'8.7',
                     '8:8':'8.8','8..8':'8.8','08.9':'8.9','8,9f':'8.9',
                     '9,.0':'9.0','9.':'9.0'}, inplace=True)
```

```python
df.score = df.score.astype(float)
```

```python
df.info()
```

### Saving the cleaned dataset

```python
df.to_csv('clean_messy_data.csv', index=False)
```
