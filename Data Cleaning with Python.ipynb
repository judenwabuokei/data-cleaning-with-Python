## Data Cleaning with Python

The dataset I'd be cleaning is the movies dataset that contains details about some movies, the year it was produced, stars, genre and gross.

```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
```

```python
df
```

```python
sum(df.MOVIES.duplicated())       #Total Number of duplicate entries
```

```python
df.drop_duplicates('MOVIES', inplace=True)     #Remove Duplicates
```

```python
df
```

```python
df.reset_index(inplace=True, drop=True)     #Resets Dataset Index
```

```python
df
```

```python
 df.columns = df.columns.str.lower().str.strip().str.replace('-','_')
```

```python
df.head()
```

```python
 df.year = df.year.str.extract('(\d{4})')        #Extracts the first 4 integers from a string.
```

```python
df
```

```python
 df.genre = df.genre.str.split(',', expand=True)[0].str[1:]
```

```python
df
```

```python
df.drop('one_line', axis=1, inplace=True)  
df.drop('(2021', axis=1, inplace=True)             #axis = 1 [drop column], axis=0 [drop row]
```

```python
df
```

```python
df.info()
```

```python
df.stars = df.stars.str.split('Stars', expand=True)[1].str.replace('\n', '').str.replace(':','')
```

```python
df
```

```python
 df.genre = df.genre.str.strip()      #stripping the column 'genre'
```

```python
df
```

```python
df.votes = df.votes.fillna(0)
```

```python
df.runtime = df.runtime.fillna(0)
```

```python
df.gross = df.gross.fillna(0)
```

```python
df.rating = df.rating.fillna(0)
```

```python
df.to_csv('cleaned_movies.csv', index=False)
```
